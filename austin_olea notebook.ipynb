```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
file_path = ('zhvi_data.csv')

data = pd.read_csv(file_path)

data.head()
```

# Florida

```python
fl_filter=data['State']=="FL"
fl_df=data[fl_filter]
display(fl_df)
```

```python
fl_percent_cols=fl_df[['RegionName','2017-03-31', '2020-02-29', '2020-03-31', '2020-12-31']]
miami_counties=["Miami-Dade County", "Broward County", "Palm Beach County", "Monroe County", "Collier County", "Lee County", "Hendry County", "Glades County", "Okeechobee County", "Martin County"]

pre_covid=[]
covid=[]

for each_county in miami_counties:
    fl_cnt_filter=fl_percent_cols['RegionName']==each_county  #just change county name and run cell
    
    mar2017=fl_percent_cols.loc[fl_cnt_filter,'2017-03-31']
    feb2020=fl_percent_cols.loc[fl_cnt_filter,'2020-02-29']
    mar2020=fl_percent_cols.loc[fl_cnt_filter,'2020-03-31']
    dec2020=fl_percent_cols.loc[fl_cnt_filter,'2020-12-31']
    
    pre_covid_annual=((((feb2020-mar2017)/feb2020)/3)*100).values[0]
    covid_annual=(((((dec2020-mar2020)/dec2020)/10)*10)*100).values[0]

    pre_covid.append(pre_covid_annual)
    covid.append(covid_annual)

display(pre_covid)
display(covid)
```
