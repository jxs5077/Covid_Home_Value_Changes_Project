```python
# Dependencies
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
file_one = "output_data/us-states.txt"
us_states = pd.read_csv(file_one)
file_two = "output_data/us-counties.txt"
us_counties = pd.read_csv(file_two)
```

```python
#Use groupby method to isolate states
grouped = us_states.groupby(['state'])
```

```python
#Calculate maximum number of cases
max_cases = grouped['cases'].max()
```

```python
#Create a data frame
remove_terr = pd.DataFrame(max_cases) 

#Remove US territories
fifty_states = remove_terr.drop(['Guam', 'Puerto Rico', 'Northern Mariana Islands', 'Virgin Islands', 'District of Columbia'])
```

```python
#Sort cases by highest to lowest number by state
fifty_states = fifty_states.sort_values(['cases'],ascending='False')
```

```python
#Create a bar chart to find top three states with highest amount of cases
panda_chart = fifty_states.plot(kind='barh', figsize=(15, 10))
panda_chart.set_xlabel("Total Number")
panda_chart.set_ylabel("States")
plt.title("States and Maximum Cases")
# plt.tight_layout()
plt.show()
```

```python
#Filter state data for California, groupby counties, find the maximum cases per county and save as df

#CA
#Filtering df at state and cases
filter_CA = us_counties.loc[us_counties['state'] == 'California', :]
filter_total1 = filter_CA.loc[filter_CA['cases'] >= 50000, :]

#Grouping by county and calculating max cases per county
grouped = filter_total1.groupby(['county'])
maxCA_cases = grouped['cases'].max()
CA_counties = pd.DataFrame(maxCA_cases)
```

```python
#TX
filter_TX = us_counties.loc[us_counties['state'] == 'Texas', :]
filter_total2 = filter_TX.loc[filter_TX['cases'] >= 50000, :]

grouped = filter_total2.groupby(['county'])
maxTX_cases = grouped['cases'].max()
TX_counties = pd.DataFrame(maxTX_cases)
```

```python
#FL
filter_FL = us_counties.loc[us_counties['state'] == 'Florida', :]
filter_total3 = filter_FL.loc[filter_FL['cases'] >= 50000, :]

grouped = filter_total3.groupby(['county'])
maxFL_cases = grouped['cases'].max()
FL_counties = pd.DataFrame(maxFL_cases)
```

```python
#Sort values in descending order for covid cases for each state df
CA_counties = CA_counties.sort_values(['cases'],ascending=True)
TX_counties = TX_counties.sort_values(['cases'], ascending=True)
FL_counties = FL_counties.sort_values(['cases'], ascending=True)
```

```python
#Create bar chart

#CA
panda_chart = CA_counties.plot(kind='barh', figsize=(15, 10), color='coral')
panda_chart.set_xlabel("Total Number")
panda_chart.set_ylabel("Counties in CA")
plt.title("California Counties with Highest Covid Cases")
# plt.tight_layout()
plt.show()
```

```python
#TX
panda_chart = TX_counties.plot(kind='barh', figsize=(15, 10), color='royalblue')
panda_chart.set_xlabel("Total Number")
panda_chart.set_ylabel("Counties in TX")
plt.title("Texas Counties with Highest Covid Cases")
# plt.tight_layout()
plt.show()
```

```python
#FL
panda_chart = FL_counties.plot(kind='barh', figsize=(15, 10), color='seagreen')
panda_chart.set_xlabel("Total Number")
panda_chart.set_ylabel("Counties in FL")
plt.title("Florida Counties with Highest Covid Cases")
# plt.tight_layout()
plt.show()
```
