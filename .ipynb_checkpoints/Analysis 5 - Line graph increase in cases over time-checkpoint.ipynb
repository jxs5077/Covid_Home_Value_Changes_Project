```python
# Dependencies
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
file_one = "output_data/us-states.txt"
us_states = pd.read_csv(file_one)
```

```python
#Change dtype of date column
us_states['date'] = pd.to_datetime(us_states['date'])
us_states['date']
```

```python
#Group date into months and find maximum amount of cases each month
month = us_states.groupby(us_states['date'].dt.strftime('%B'))['cases'].max().sort_values()
month
```

```python
#Plot increase in cases over time
months = list(month.index)
cases = month.values

plt.figure(figsize=(12,8))
plt.plot(months,cases, marker="^", color="red")
plt.xlabel("Month", fontsize=10)
plt.ylabel("Maximum Positive Cases")
plt.title("COVID-19 - Maximum Positive Cases Feb2020-Jan2021")
plt.xticks(fontsize=8,rotation=90)

plt.show()
```
