```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read raw data

file_path = "zillow_data/zhvi_data/County_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv"
home_values_df=pd.read_csv(file_path)

home_values_df.head()
```

```python
# Find Column names
year_list=list(home_values_df.columns)
year_list
```

```python
# Create new df with only interested columns (location info, and 2017-2020 data)
interested_columns=['RegionName', 'RegionType','State', 'Metro', 'StateCodeFIPS', 'MunicipalCodeFIPS', '2017-01-31', '2017-02-28', '2017-03-31', '2017-04-30', '2017-05-31', '2017-06-30', '2017-07-31',
       '2017-08-31', '2017-09-30', '2017-10-31', '2017-11-30', '2017-12-31', '2018-01-31', '2018-02-28', '2018-03-31', '2018-04-30', '2018-05-31', '2018-06-30', '2018-07-31',
       '2018-08-31', '2018-09-30', '2018-10-31', '2018-11-30', '2018-12-31', '2019-01-31', '2019-02-28', '2019-03-31', '2019-04-30', '2019-05-31', '2019-06-30', '2019-07-31',
       '2019-08-31', '2019-09-30', '2019-10-31', '2019-11-30', '2019-12-31', '2020-01-31', '2020-02-29','2020-03-31', '2020-04-30', '2020-05-31', '2020-06-30', '2020-07-31',
       '2020-08-31', '2020-09-30', '2020-10-31', '2020-11-30', '2020-12-31']

values_df=home_values_df.loc[:, interested_columns]

values_df.head()
```

```python
# Checking data for missing values
values_df.count()
```

```python
# check how many null values are in the numerical columns 
values_clean_df=values_df.loc[:,'2017-01-31':'2020-12-31'].dropna(how="any")
values_clean_df.count()
```

```python
# Looking at region name for duplicates
values_df.groupby('RegionName').count()
```
